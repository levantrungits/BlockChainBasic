```python
# 0. Import Lib For Hash Function
import hashlib as hasher
from datetime import datetime
```

```python
# 1. Defined A Block
class Block:
    def __init__(self, index, timestamp, data, previous_hash):
        self.index = index
        self.timestamp = timestamp
        self.data = data
        self.previous_hash = previous_hash
        self.hash = self.hash_block()
        
    def hash_block(self):
        sha = hasher.sha256()
        strHash = str(self.index) + str(self.timestamp) + str(self.data) + str(self.previous_hash)
        sha.update(strHash.encode("utf-8"))
        return sha.hexdigest()
```

```python
# 2. Create Root Block By Hand - Index = 0
def create_genesis_block():
    return Block(0, datetime.now(), "Genesis Block", "0")
```

```python
# 3. Craete New Block From Old Block
def next_block(last_block):
    this_index = last_block.index + 1
    this_timestamp = datetime.now()
    this_data = "Hello world! I'm block" + str(this_index)
    this_hash = last_block.hash
    return Block(this_index, this_timestamp, this_data, this_hash)
```

```python
# 4. Add Many Blocks Into Blockchain By "For" Function
blockchain = [create_genesis_block()]
previous_block = blockchain[0]

num_of_blocks_to_add = 20

for idx in range(0, num_of_blocks_to_add):
    block_to_add = next_block(previous_block)
    blockchain.append(block_to_add)
    previous_block = block_to_add
    #print("Block #{} đã được thêm vào Blockchain".format(block_to_add.index))
    
for idx in blockchain:
    print(idx.index)
    print(idx.data)
    print(idx.previous_hash)
    print(idx.hash)
    print("\n")
```
